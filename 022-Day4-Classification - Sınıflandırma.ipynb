# Feature Enginering

```python
df['Attrition']
```

```python
# Dictionary Sözlük
#Kelime: Açıklama
# d kısaltma
```

```python
d={"Yes":1,"No":0} #işten ayrılanlar=1 ayrılmayan=0
```

```python
df['Attrition']=df['Attrition'].map(d)
```

```python
df['Attrition']
```

```python
bos=df[df['Attrition'].isnull()]

# Attrition ları boş olanları bul
```

```python
dolu=df[df['Attrition'].notnull()]

# dolu olanları bulacak - yukaridaki kod ie birlikte dolu ve boşları ayırdı
```

```python
x=dolu.drop('Attrition',axis=1)
y=dolu['Attrition'] 

#35 sütun ismi yazmadan dolunun içerisindeki Attrition  sütununu sil
#axis=1  sütun siler
```

```python
x.sample(5)
#attrition yok. drop ettik sildik
```

```python
x.shape
```

```python
x['BusinessTravel'].value_counts()
```

```python
x=pd.get_dummies(x,drop_first=True)

#yazıları rakama çevirdik
```

```python
x.head()
```

```python
x.shape

#her yazı olan değerler  için ilave sütunlar ilave etti. ve altına 1-0 yazdı
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
g=GaussianNB()
b=BernoulliNB()

#classification algoritmaları
```

```python
g=GaussianNB()
b=BernoulliNB()
```

```python
g.fit(x_train,y_train)
```

```python
tahmin1=g.predict(x_test)
```

```python
b.fit(x_train,y_train)
```

```python
tahmin2=b.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report
```

```python
accuracy_score(tahmin1,y_test)
```

```python
accuracy_score(tahmin2,y_test)
```

```python
print(classification_report(tahmin2,y_test))
```

```python
confusion_matrix(tahmin2,y_test)
```

```python
import seaborn as sns
```

```python
sns.heatmap(confusion_matrix(tahmin2,y_test),annot=True)
```
